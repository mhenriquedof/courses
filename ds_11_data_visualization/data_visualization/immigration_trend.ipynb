```python
df_south_america = df.query("Região == 'América do Sul'")
df_south_america = df_south_america.sort_values(by='Total')
df_south_america
```

```python
colors = []

for country in df_south_america.index:
    if country == 'Brasil':
        colors.append('green')
    else:
        colors.append('silver')
```

```python
fig, ax = plt.subplots(figsize=(12,7))

ax.barh(df_south_america.index, df_south_america['Total'], color=colors)
ax.set_title('Immigration from South America to Canada - Brazil was the 4th country\n1980 to 2013', loc='left', fontsize=18)
ax.set_xlabel('Immigrations', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(df_south_america['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('immigration_southamerica_canada.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
ax = sns.barplot(data=df_top_immigrations, x='Total', y=df_top_immigrations.index)
ax.set(title="Top 10 countries with immigrations to Canada",
       xlabel='Immigrations',
       ylabel='')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=df_top_immigrations, x='Total', y=df_top_immigrations.index)
ax.set_title('Top 10 immigration to Canada', loc='left', fontsize=16)
ax.set_xlabel('Immigrations', fontsize=14)
ax.set_ylabel('')

plt.show()
```

```python
def gen_graph_palette(palette):
    fig, ax = plt.subplots(figsize=(12,7))
    
    ax = sns.barplot(data=df_top_immigrations, x='Total', y=df_top_immigrations.index, palette=palette)
    ax.set_title('Immigration from Brazil to Canada\n1980 to 2013', loc='left', fontsize=18)
    ax.set_xlabel('Immigrations', fontsize=14)
    ax.set_ylabel('')

    plt.show()
```

```python
gen_graph_palette('Blues_r')
```
