```python
import pandas as pd
```

```python
data = pd.read_csv('data/pistachio.csv')
data
```

```python
data.isna().sum().sum()
```

```python
data['Class'] = data['Class'].replace({'Kirmizi_Pistachio': 0, 'Siit_Pistachio': 1})
data.info()
```

```python
x = data.drop(columns='Class')
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=42)
```

```python
corr_data = pd.concat([x_test, y_test], axis=1).corr()
corr_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_data, annot=True, cmap='coolwarm', fmt='.2f')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(random_state=42)
rfc.fit(x_train, y_train)
```

```python
importancia = pd.DataFrame({'colunas': x_train.columns, 'importancia': rfc.feature_importances_})
best_importancias = importancia.sort_values(by='importancia', ascending=False).head(20)
best_importancias
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(best_importancias, x='colunas', y='importancia')
ax.set_xticklabels(ax.get_xticklabels(), rotation=75);
```

---

```python
from sklearn.feature_selection import SelectFromModel
```

```python
f_selection_model = SelectFromModel(rfc, max_features=10, prefit=True)
```

```python
x_train_select = f_selection_model.transform(x_train)
x_test_select = f_selection_model.transform(x_test)
```

```python
rfc_select = RandomForestClassifier(n_estimators=20, random_state=2)
```

```python
rfc_select.fit(x_train_select, y_train)
```

```python
rfc_select.score(x_test_select, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

rfc_grid = RandomForestClassifier(random_state=2)

parametros_busca = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid_search = GridSearchCV(estimator=rfc_grid, param_grid=parametros_busca,
                           scoring='accuracy', cv=5, n_jobs=-1)

grid_search.fit(x_train, y_train)
```

```python
f_selecionadas = SelectFromModel(grid_search.best_estimator_, max_features=10,prefit=True)
```

```python
x_train_selected = f_selecionadas.transform(x_train)
x_test_selected = f_selecionadas.transform(x_test)
```

```python
rfc_select_grid = grid_search.best_estimator_
rfc_select_grid.fit(x_train_selected, y_train)
```

```python
rfc_select_grid.score(x_test_selected, y_test)
```

---

```python
rfc_rfecv = RandomForestClassifier(n_estimators=20, random_state=2)
```

```python
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
rfecv = RFECV(
    estimator=rfc_rfecv,
    step=1,
    cv=cv,
    scoring='roc_auc', #roc_auc
    min_features_to_select=1,
    importance_getter='auto'
)
```

```python
rfecv.fit(x_train, y_train)
```

```python
rfecv.n_features_
```

```python
rfecv.ranking_
```

```python
y_pred = rfecv.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
matriz_confusao = confusion_matrix(y_test, y_pred)

sns.heatmap(matriz_confusao, annot=True, fmt='d')
```
